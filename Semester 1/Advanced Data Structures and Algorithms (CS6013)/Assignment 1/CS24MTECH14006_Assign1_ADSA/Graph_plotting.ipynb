```python
# Randomized quick sort graph plotting
import matplotlib.pyplot as plt # Importing plotting library for visualizing the data


# Function for reading data from the text file & parse it into arrays
def read_data_from_file(filename):
    sizes_of_arrays = []
    average_no_comparisons = []

    with open(filename, 'r') as file:
        for line in file:
            if "Array size" in line:
                parts = line.split(',')
                size = int(parts[0].split(':')[1].strip())
                comps_avg = float(parts[1].split(':')[1].strip())
                
                sizes_of_arrays.append(size)
                average_no_comparisons.append(comps_avg)
    
    return sizes_of_arrays, average_no_comparisons

# Reading data fro 'average_comparisons.txt' file
filename = 'average_comparisons.txt'
sizes_of_arrays, average_no_comparisons = read_data_from_file(filename)

# Converting all comparisons to lakhs (1 lakh = 1,00,000)
comparision_average_in_lakhs = [comp / 1e5 for comp in average_no_comparisons]

# Converting array sizes to more readable format
labels_array_size_x_axis = ['10^4', '10^5', '10^6', '10^7']

# Plotting graph
plt.figure(figsize=(8, 6))
plt.bar(labels_array_size_x_axis, comparision_average_in_lakhs, color='blue')

# Adding labels & title
plt.xlabel('Array Size')
plt.ylabel('Avg Comparisons (Lakhs)')
plt.title('Randomized Quick Sort-Average Comparisons (Lakhs) vs Array Size ')

# For better readability, add values above bars
for i, v in enumerate(comparision_average_in_lakhs):
    plt.text(i, v + 0.1, f'{v:.2f}', ha='center', va='bottom')

plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Showing plot
plt.show()
```

```python
# Red black tree insetion 


# Define file names
files = [
    'average_rotations_height_size10000.txt',
    'average_rotations_height_size100000.txt',
    'average_rotations_height_size1000000.txt',
    'average_rotations_height_size10000000.txt'
]

# Define array sizes and their corresponding labels
array_sizes = [10000, 100000, 1000000, 10000000]
array_size_labels = ['10^4', '10^5', '10^6', '10^7']

# Initialize lists for storing the values
average_rotations = []
average_heights = []

# Function to read and parse data from files
def parse_file(filename):
    with open(filename, 'r') as f:
        lines = f.readlines()
        rotations = float(lines[0].split(":")[1].strip())  # Extract rotations
        height = float(lines[1].split(":")[1].strip())  # Extract height
    return rotations, height

# Read data from each file
for file in files:
    rotations, height = parse_file(file)
    average_rotations.append(rotations / 100000)  # Convert to lakhs
    average_heights.append(height)

# Plot Average Rotations vs Size of Array (in lakhs)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.bar(array_size_labels, average_rotations, color='blue')
plt.xlabel('Size of Array')
plt.ylabel('Average Rotations (in lakhs)')
plt.title('Average Rotations vs Size of Array (in lakhs)')
plt.ylim(0, max(average_rotations) * 1.1)  # Scale y-axis based on max rotations

# Plot Average Height vs Size of Array
plt.subplot(1, 2, 2)
plt.bar(array_size_labels, average_heights, color='skyblue')
plt.xlabel('Size of Array')
plt.ylabel('Average Height')
plt.title('Average Height vs Size of Array')
plt.ylim(0, 40)  # Scale y-axis from 0 to 40 as per instruction

# Display the plots
plt.tight_layout()
plt.show()
```

```python
# AVL tree insetion 

# Define file names for rotations and height
rotation_files = [
    'avl_rotations_data_size10000.txt',
    'avl_rotations_data_size100000.txt',
    'avl_rotations_data_size1000000.txt',
    'avl_rotations_data_size10000000.txt'
]

height_files = [
    'avl_height_data_size10000.txt',
    'avl_height_data_size100000.txt',
    'avl_height_data_size1000000.txt',
    'avl_height_data_size10000000.txt'
]

# Define array sizes and corresponding labels
array_sizes = [10000, 100000, 1000000, 10000000]
array_size_labels = ['10^4', '10^5', '10^6', '10^7']

# Initialize lists for storing values
average_rotations = []
average_heights = []

# Function to read rotations data from files
def parse_rotations_file(filename):
    with open(filename, 'r') as f:
        lines = f.readlines()
        rotations = float(lines[0].split(":")[1].strip())  # Extract rotations
    return rotations

# Function to read height data from files
def parse_height_file(filename):
    with open(filename, 'r') as f:
        lines = f.readlines()
        height = float(lines[0].split(":")[1].strip())  # Extract height
    return height

# Read rotations data from each file
for file in rotation_files:
    rotations = parse_rotations_file(file)
    average_rotations.append(rotations / 100000)  # Convert to lakhs

# Read height data from each file
for file in height_files:
    height = parse_height_file(file)
    average_heights.append(height)

# Plot Average Rotations vs Size of Array (in lakhs)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.bar(array_size_labels, average_rotations, color='blue')
plt.xlabel('Size of Array')
plt.ylabel('Average Rotations (in lakhs)')
plt.title('Average Rotations vs Size of Array (in lakhs)')
plt.ylim(0, 80)  # Scale y-axis from 0 to 80 lakh

# Plot Average Height vs Size of Array
plt.subplot(1, 2, 2)
plt.bar(array_size_labels, average_heights, color='skyblue')
plt.xlabel('Size of Array')
plt.ylabel('Average Height')
plt.title('Average Height vs Size of Array')
plt.ylim(0, 40)  # Scale y-axis from 0 to 40 as per instruction

# Display the plots
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Defining file names for rotations & height after rotations
deletion_files_avl_tree = [
    'avl_deletion_data_size10000.txt',
    'avl_deletion_data_size100000.txt',
    'avl_deletion_data_size1000000.txt',
    'avl_deletion_data_size10000000.txt'
]

# Defining array sizes and corresponding labels
array_sizes = [10000, 100000, 1000000, 10000000]
array_size_labels = ['10^4', '10^5', '10^6', '10^7']

# Initialize lists for storing values
average_rotations = []
average_heights_after_rotations = []

# Function to read rotations and height after rotations from files
def parse_deletion_file(filename):
    with open(filename, 'r') as f:
        lines = f.readlines()
        rotations = float(lines[0].split(":")[1].strip()) 
        height_after_deletion = float(lines[1].split(":")[1].strip())  
    return rotations, height_after_deletion

# Read rotations and height data from each file
for file in deletion_files_avl_tree:
    rotations, height_after_deletion = parse_deletion_file(file)
    average_rotations.append(rotations / 100000)  # Convert rotations to lakhs
    average_heights_after_rotations.append(height_after_deletion)

# Plot Average rotations vs Size of Array (in lakhs)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.bar(array_size_labels, average_rotations, color='blue')  
plt.xlabel('Size of Array')
plt.ylabel('Average rotations (in lakhs)')
plt.title('Average rotations vs Size of Array (in lakhs)')
plt.ylim(0, 110)  # Scale y-axis from 0 to 110 lakh

# Plot Average Height After rotations vs Size of Array
plt.subplot(1, 2, 2)
plt.bar(array_size_labels, average_heights_after_rotations, color='skyblue')  
plt.xlabel('Size of Array')
plt.ylabel('Average Height After rotations')
plt.title('Average Height After rotations vs Size of Array')
plt.ylim(0, 40)  # Scale y-axis from 0 to 40 as per instruction

# Display the plots
plt.tight_layout()
plt.show()
```
