```python
# Question number 4 a)
#importing libraries
import numpy as num
from sklearn.metrics import accuracy_score #For commputing accu by comparing true & predicted labels
from sklearn.svm import SVC #this is to perform classification task


# Dataset to be loaded here
test_feature_data = num.loadtxt('http://www.amlbook.com/data/zip/features.test')
train_feature_data = num.loadtxt('http://www.amlbook.com/data/zip/features.train')


# Data filtering for particular labels 1 or 5
train_feature_data = train_feature_data[(train_feature_data[:, 0] == 1) | (train_feature_data[:, 0] == 5)]
test_feature_data = test_feature_data[(test_feature_data[:, 0] == 1) | (test_feature_data[:, 0] == 5)]

# For separating the labels and features

# Extract the features from column 2 & column 1
X_feature_training  = train_feature_data[:, 1:3]
X_feature_testing =test_feature_data[:, 1:3]

# Modifing the target label to binary ( 1 for the digit '1' &  -1 for digit the '5')
y_label_training = num.where(train_feature_data[:, 0] == 1, 1, -1)
y_label_testing = num.where(test_feature_data[:, 0] == 1, 1, -1)





# Train SVM using the linear kernel
linear_kernel = SVC(kernel='linear', C=1.0)  #  The value of C is 1.0 for S V M with soft margin 
linear_kernel.fit(X_feature_training , y_label_training )

# Now make pred on the test set by trained model
y_label_predict = linear_kernel.predict(X_feature_testing )

# calculating the  no. of support vectors & accuracy
number_s_v_4a= len(linear_kernel.support_)
accu_svm_4a =accuracy_score(y_label_testing, y_label_predict)
print(f'Printing answer for question 4)a')
print(f'Accuracy is { accu_svm_4a }')
print(f'No. of Support Vectors are {number_s_v_4a}')





##******Que 4) b ******
print(f'\n\n\nPrinting answer for question 4) b')
# Defining sample sizes to use

sample_size_train = [50, 100, 200, 800]

# Loop for each sample size above
for s in sample_size_train:
    print(f"\nTraining with first {s} samples- ")
    
    # Training svm using linear kernel on subset of training data
    s_v_m = SVC(kernel='linear', C=1.0)
    s_v_m.fit(X_feature_training [:s], y_label_training [:s])
    
    # Making predictions on testing data
    y_label_predict = s_v_m.predict(X_feature_testing )
    
    # Calculating accuracy on testing data
    accu_svm_4b = accuracy_score(y_label_testing, y_label_predict)
    
    # Getting no. of sv used in model.
    number_sv_4b= len(s_v_m.support_)
    
    # For the cuurent size of sample, results will be printed
    print(f"Sample Size is {s}")
    print(f"Accuracy is {accu_svm_4b}")
    print(f"No. of Support Vectors are {number_sv_4b}")




##******Que 4) c ******
print(f'\n\n\nPrinting answer for question 4) c')
# Defining values of C parameter &  the polynomial degrees
# Looping over each C parameter value & every polynomial degree

for C_value in [0.0001, 0.001, 0.01, 1]:

    for degree in [2, 5]:
                
        # Training model with the polynomial kernel using particular parameter and degree
        svm_poly = SVC(kernel='poly', degree=degree, C=C_value)
        svm_poly.fit(X_feature_training , y_label_training )

        # Making the predictions data sets
        y_pred_poly_train = svm_poly.predict(X_feature_training)
        accuracy_poly_train = accuracy_score(y_label_training , y_pred_poly_train)

        # Finding the no. of sv
        support_vectors_poly = svm_poly.support_vectors_.shape[0]

        # Calculating the training and testing accu
        testing_error_poly = 1 - accuracy_score(y_label_testing, svm_poly.predict(X_feature_testing ))
        training_error_poly =1-accuracy_poly_train

        
        print(f"For C={C_value} & Degree={degree}:")
        print(f"Training Error is {training_error_poly:.4f}.")
        print(f"Testing error is {testing_error_poly :.4f}.")
        print(f"No. of support vectors are {support_vectors_poly}.\n")





##******Que 4) d ******
print(f'\n\n\nPrinting answer for question 4) d')
# Define values of C parameter
C_parameter_value = [0.01, 1, 100, 1e4, 1e6]

# Loop over each and every value of C
for C in C_parameter_value:
    print(f"\nEvaluating for C = {C}")
    
    # Training SVM model by the RBF kernel
    rbf_kernel_svm = SVC(kernel='rbf', C=C, gamma='scale')  # 'scale' is default for RBF kernel in the scikit-learn
    rbf_kernel_svm.fit(X_feature_training, y_label_training)
    
    # Making predictions
    y_label_traininging_pred = rbf_kernel_svm.predict(X_feature_training)
    y_label_testinging_pred = rbf_kernel_svm.predict(X_feature_testing)
    
    # Calculating accuracy
    accu_train = accuracy_score(y_label_training, y_label_traininging_pred)
    accu_test = accuracy_score(y_label_testing, y_label_testinging_pred)
    
    # Calculating testing error & training error
    error_testing = 1 - accu_test
    error_training = 1 - accu_train
    
    # Displaying obtained results
    print(f"Training error- {error_training}")
    print(f"Test error- {error_testing}")


```

```python
# Question number 5)
import numpy as num
from sklearn.svm import SVC # importing support vector classifier
from sklearn.metrics import accuracy_score #For measuring prediction accuracy

# Function for loading the data from text files
def dataset_loading_f(training_data_path, training_labels_path, validat_data_path, validat_labels_path):

    #Loading training data
    training_data = num.loadtxt(training_data_path)
    #Loading training labels
    training_label = num.loadtxt(training_labels_path).astype(int)
    #Loading test data
    validat_data = num.loadtxt(validat_data_path)
    #Loading test labels
    validat_label = num.loadtxt(validat_labels_path).astype(int)
    
    return training_data, training_label, validat_data, validat_label

# Function with the help of Linear Kernel for Support vector machine
def linear_kernel_s_v_m(X_training, y_training, X_testing, y_testing):
    linear_model = SVC(kernel='linear', C=1.0) #penalty parameter c=1.0
    # Now train model on traning data with linear model
    linear_model.fit(X_training, y_training)

    #For evaluating training performance, predicting labels
    training_linear_prediction = linear_model.predict(X_training)
    testing_prediction_linear= linear_model.predict(X_testing)

    #calculating accuracy on testing & traninig dataset
    accuracy_training_linear = accuracy_score(y_training, training_linear_prediction)
    accuracy_testing_linear = accuracy_score(y_testing, testing_prediction_linear)
    #calculating accuracy on testing & traninig dataset
    error_training_linear = 1 - accuracy_training_linear
    error_test_linear = 1 - accuracy_testing_linear
    #Finding no. of support vectors
    no_vectors_linear = len(linear_model.support_)

    #Displayin output
    print(f'***Linear Support Vector Classifier***')
    print(f'Training Error for linear Support Vector Classifier is {error_training_linear:.5f}.')
    print(f'Testing Error  for linear Support Vector Classifier is {error_test_linear:.5f}.')
    print(f'No. of Support Vectors are {no_vectors_linear}.')

    return error_training_linear, error_test_linear, no_vectors_linear

# Function with the help of RBF Kernel for Support vector machine
def rbf_kenel_s_v_m(X_training, y_training, X_testing, y_testing):
   #Creating model using RBF kernel with parameter = 1.0 & value of gamma is 0.001
    rbf_model = SVC(kernel='rbf', C=1.0, gamma=0.001)
    #Using training data, model is trained
    rbf_model.fit(X_training, y_training)

    #Predicting labels on testing & training dataset.
    training_rbf_prediction = rbf_model.predict(X_training)
    testing_rbf_prediction = rbf_model.predict(X_testing)

    #Calculating accu on testing & training dataset
    accuracy_training_rbf = accuracy_score(y_training, training_rbf_prediction)
    accuracy_testing_rbf = accuracy_score(y_testing, testing_rbf_prediction)
    
    # Calculating error 
    error_training_rbf = 1 - accuracy_training_rbf
    error_testing_rbf = 1 - accuracy_testing_rbf

    #Finding no. of support vectors in RBF model
    no_vectors_rbf = len(rbf_model.support_)

    #Displaying the result
    print(f'***RBF Support Vector Classifier***:')
    print(f'Training Error for RBF Support Vector Classifier is {error_training_rbf:.5f}.')
    print(f'Testing Error for RBF Support Vector Classifier is {error_testing_rbf:.5f}.')
    print(f'No. of Support Vectors are {no_vectors_rbf}.')

    return error_training_rbf, error_testing_rbf, no_vectors_rbf

#  Function with the help of polynomial Kernel for Support vector machine
def polynomial_kernel_s_v_m(X_training, y_training, X_testing, y_testing):
    #Creating poly model having degree 2, C 1 & coef0 1
    polynomial_model = SVC(kernel='poly', degree=2, coef0=1, C=1.0)
    #Training model on traning dataset
    polynomial_model.fit(X_training, y_training)

    #Predicting labels on training & validation data(test data)
    accuracy_prediction_training =  polynomial_model.predict(X_training)
    accuracy_prediction_testing =  polynomial_model.predict(X_testing)

    #Calculating accu on testing & training data
    accuracy_training_poly = accuracy_score(y_training, accuracy_prediction_training)
    accuracy_testing_poly = accuracy_score(y_testing, accuracy_prediction_testing)

    #Finding error
    error_training_poly = 1 - accuracy_training_poly
    error_testing_poly = 1 - accuracy_testing_poly
    #Obtain no. of support vectors used in poly svm 
    no_vectors_poly = len( polynomial_model.support_)

    #Displaying the result obatained
    print(f'Polynomial SVC (degree=2):')
    print(f'Training Error for polynomial Support Vector Classifier is{error_training_poly:.5f}.')
    print(f'Testing Error for polynomial Support Vector Classifier is{error_testing_poly:.5f}.')
    print(f'No. of Support Vectors are {no_vectors_poly}.')

    return error_training_poly, error_testing_poly, no_vectors_poly

# Main function to load data and run the different SVC models
def main():
    # Loading  GISETTE dataset from particular files mentioned
    print("\n Loading the data...........")
    training_data, training_label, validat_data, validat_label = dataset_loading_f(
        'gisette_train.data', 
        'gisette_train.labels', 
        'gisette_valid.data', 
        'gisette_valid.labels'
    )

    # For testing & training data is assigned here
    X_training = training_data
    y_training = training_label
    X_testing = validat_data
    y_testing = validat_label

    # Run linear model & display results for Linear SVC 
    print("\nRunning Linear SVC ....")
    linear_kernel_s_v_m(X_training, y_training, X_testing, y_testing)

    # Run RBF model & display results for RBF SVC
    print("\nRunning RBF SVC ....")
    rbf_kenel_s_v_m(X_training, y_training, X_testing, y_testing)

    # Run polynomial model & display results for Polynomial SVC
    print("\nRunning Polynomial SVC (degree=2) ....")
    polynomial_kernel_s_v_m(X_training, y_training, X_testing, y_testing)

# Calling the main function
if __name__ == "__main__":
    main()
```
